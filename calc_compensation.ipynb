## ___Розрахунок компенсації громадянам втрати частини грошових доходів у зв'язку з порушенням термінів їх виплати___

Розрахунок здійснюється у відповідності до Порядку проведення компенсації громадянам втрати частини грошових доходів у зв'язку з порушенням термінів їх виплати, затвердженого постановою Кабінету Міністрів України №159 від 21.02.1995

[Посилання на порядок](https://zakon.rada.gov.ua/laws/show/159-2001-%D0%BF#Text)

```python
# Імпорт відповідних бібліотек для проведення розархунку
import numpy as np
import pandas as pd
```

1. Створення датафрейму до якого включаються індекси споживчих цін за відповідні періоди коли мало місце порушення термінів виплати грошових доходів.
    - Індекси споживчих цін та індекс інфляції є тотожними поняттями у вказаному розархунку
    
    [Посилання на сайт містату де розміщені дані про індекси споживчих цін за відповідні періоди](https://ukrstat.gov.ua/operativ/menu/menu_u/cit.htm)

```python
# Створення датафрейму
df =pd.DataFrame(data={'місяці':['січень', 'лютий', 'березень', 'квітень', 'травень', 'червень', 'липень', 'серпень', 'вересень', 'жовтень', 'листопад', 'грудень'],
                                '2022':[1.013, 1.016, 1.045, 1.031, 1.027, 1.031, 1.007, 1.011, 1.019, 1.025, 1.007, 1.007],
                                '2023':[1.008, 1.007, 1.015, 1.002, 1.005, 1.008, 0, 0, 0, 0, 0, 0]})
```

```python
df
```

```python
# Присвоєння змінній даних про розмір доходу, який повинен бути виплачений
# В даному прикладі дохід є незмінним на протязі всього періоду не виплати
income = 22199.33
```

2. Обробка даних зі створеного датафрейму з метою об'єднання даних про індекси споживчих цін в єдиний датафрейму за відповідні місяці років у яких мало місце порушення строків виплати доходу

```python
# Опрацювання датафрейму з метою створення окремого датафрейму для подальшого проведення розрахунку
df_2022=df['2022'][1:]
df_2023=df['2023'][:6]
df_month_2022 = df['місяці'][1:12]
df_month_2023 = df['місяці'][:6]
df_monthes=pd.concat([df_month_2022,df_month_2023])
df_monthes=df_monthes.reset_index(drop=True)
df_indexes= pd.concat([df_2022,df_2023])
df_indexes=df_indexes.reset_index(drop=True)
numbers_range = range(1, 18)
month_list = list(numbers_range)
result_df = pd.concat([df_monthes, df_indexes], axis=1)
result_df.columns = ['місяці', 'індекси']
```

```python

result_df
```

3. Проведення розрахунку та виведення результату

```python
values=[]
```

```python
for index, i in result_df.iterrows():
    month=i['місяці']
    print(f'{index} Розрахунок для місяця {month}\n')
    value =result_df['індекси'][index+1:].prod()       
    value=value*100
    value=round(value,2)
    print(f'Індекс споживчих цін за період невиплати грошового доходу -> {value}')
    value=value-100
    value=round(value,2)
    print(f'Величина приросту індексу споживчих цін -> {value}')
    value=value*income/100
    value=round(value,2)
    values.append(value)
    
    print(f'Сума компенсації -> {value}\n')

values=sum(values)
values=round(values,2)
print(f'Загальна сума компенсації -> {values}\n')

```
